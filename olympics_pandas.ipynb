# FUNCIONES 

```python
# Esta función cálcula la cantidad de medallas 

# input --> tipo: puede ser tipo Gold.1, Silver.1 etc.. o Combined total   
def medallas(tipo):
    return olympics_data[tipo]
```

```python
# Esta función da un puntaje según el peso de cada medalla 
# ORO = 5 puntos, PLATA = 3 puntos, BRONCE = 1 punto 

# inputs --> TipoG: de tipo Gold (Gold.1,Gold.2)
#            TipoS: de tipo Silver
#            TipoB: de tipo Bronze
def pesos(tipoG,tipoS,tipoB):
    olympics_data['Points']= (olympics_data[tipoG]*3+olympics_data[tipoS]*2+olympics_data[tipoB])
    return olympics_data['Points']
```

```python
# Esta función divide la cantidad de medallas totales por la población en un año determinado. 

# input --> year: 1960 ,1980 etc... o población promedio 
def win_pop(year):
    return (olympics_data['Combined total']/POP_data[year])
```

```python
# Esta función divide la cantidad de medallas totales por el GDP en un año determinado. 

# input --> year: 1960 ,1980 etc... o población promedio 
def win_GDP(year):
    return (olympics_data['Combined total']/GDP[year])
```
